# 3. A description of final product.

My final project is about the massive divide in Name, Image, and likeness (NIL) earnings and spending both among the players and across conferences.I used a data set that gave accurate evaluations for the top 100 college football players. From this data set, I created a Lorenz curve that was about the extreme inequality in NIL earnings. The Lorenz curve showed that a small percentage of players were earning the majority of NIL money, with another graphic showing that the top 4 players made the as the bottom 60. Although we were only able to look at the top 100, my research only lead me to believe that the inequality  gets worse as you go down the list beyond those players. I also looked at the conference level spending with NIL which proved the power conferences spend over 10 times more on NIL than the smaller conferences. Overall, my final project is meant to show people that the college football landscape is being completely reshaped by NIL, with the benefits being concentrated in the hands of a few elite players and programs.
